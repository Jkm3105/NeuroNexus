```python
correlation = df.corr()
plt.figure(figsize=(5,5))
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrix")
plt.show()
```

```python
varaince = df.var()
varaince
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df.drop(columns=['PassengerId','Pclass','Survived','SibSp','Parch']),orient='h')
plt.title("Box Plots")
plt.show()
```

```python
cfeatures=['Sex','Embarked']
for feature in cfeatures:
  plt.figure(figsize=(10,6))
  sns.countplot(data=df,x=feature,palette="Set1")
  plt.title(f'Count Plot of {feature}')
  plt.show()
```
